```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])         #(size in 1000 square feet)
y_train = np.array([250, 300, 480, 430, 630, 730])         #(price in 1000s of dollars)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x_train, y_train, marker='x')
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price (1000 USD)")
plt.title("Training data")
plt.grid(True)
plt.show()
```

```python
def compute_cost(x, y, w, b):
    """
    Computes the cost function for linear regression
    
    Args:
      x (ndarray (m,)): Data
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters  
    
    Returns:
      total_cost (float)
    """
    m = x.shape[0]
    cost_sum = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum += cost
        
    total_cost = (1 / (2 * m)) * cost_sum
    return total_cost
```

```python
w_test = 200
b_test = 50

cost = compute_cost(x_train, y_train, w_test, b_test)
print(f"Cost with w={w_test}, b={b_test}: {cost:.2f}")
```

```python
def plot_model(x, y, w, b):
    x_line = np.linspace(min(x), max(x), 100)
    y_line = w * x_line + b
    
    plt.figure(figsize=(6,4))
    plt.scatter(x, y, marker='x', label="Training data")
    plt.plot(x_line, y_line, label=f"Model: y={w}x+{b}")
    plt.xlabel("Size (1000 sqft)")
    plt.ylabel("Price (1000 USD)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_model(x_train, y_train, w=200, b=50)
```

```python
w_values = np.linspace(0, 400, 100)
cost_values = [compute_cost(x_train, y_train, w, 0) for w in w_values]

plt.figure(figsize=(6,4))
plt.plot(w_values, cost_values)
plt.xlabel("w")
plt.ylabel("Cost")
plt.title("Cost vs w (b = 0)")
plt.grid(True)
plt.show()
```
