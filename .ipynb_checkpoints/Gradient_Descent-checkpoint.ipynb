```python
import math
import copy
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10
```

```python
# Load our data set
x_train = np.array([1.0, 2.0])   #features
y_train = np.array([300.0, 500.0])   #target value
```

```python
def compute_cost(x, y, w, b):
    """
    Computes the cost function for linear regression.
    
    Args:
        x (ndarray (m,)): Data, m examples
        y (ndarray (m,)): Target values
        w, b (scalar): Model parameters
    
    Returns:
        total_cost (float): The cost of using w,b as parameters
    """
    m = x.shape[0]
    cost = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i])**2
    total_cost = 1 / (2 * m) * cost
    
    return total_cost
```

```python
def compute_gradient(x, y, w, b):
    """
    Computes the gradient for linear regression.
    
    Args:
        x (ndarray (m,)): Data, m examples
        y (ndarray (m,)): Target values
        w, b (scalar): Model parameters
    
    Returns:
        dj_dw (scalar): The gradient of the cost w.r.t. parameter w
        dj_db (scalar): The gradient of the cost w.r.t. parameter b
    """
    # Number of training examples
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb - y[i]
        dj_db += dj_db_i
        dj_dw += dj_dw_i
    
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, gradient_function):
    """
    Performs gradient descent to fit w, b.
    
    Args:
        x (ndarray (m,)): Data, m examples
        y (ndarray (m,)): Target values
        w_in, b_in (scalar): Initial values of model parameters
        alpha (float): Learning rate
        num_iters (int): Number of iterations to run gradient descent
        cost_function: Function to compute cost
        gradient_function: Function to compute gradient
    
    Returns:
        w (scalar): Updated value of parameter w
        b (scalar): Updated value of parameter b
        J_history (list): History of cost values
        p_history (list): History of parameters [w, b]
    """
    # Arrays to store cost J and w's at each iteration for graphing
    J_history = []
    p_history = []
    b = b_in
    w = w_in
    
    for i in range(num_iters):
        # Calculate the gradient and update the parameters
        dj_dw, dj_db = gradient_function(x, y, w, b)
        
        # Update parameters
        b = b - alpha * dj_db
        w = w - alpha * dj_dw
        
        # Save cost J at each iteration
        if i < 100000:  # prevent resource exhaustion
            J_history.append(cost_function(x, y, w, b))
            p_history.append([w, b])
        
        # Print cost every at intervals 10 times or as many iterations if < 10
        if i % math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
    
    return w, b, J_history, p_history
```

```python
# Initialize parameters
w_init = 0
b_init = 0
```

```python

# Gradient descent settings
iterations = 10000
tmp_alpha = 1.0e-2

```

```python
# Run gradient descent
w_final, b_final, J_hist, p_hist = gradient_descent(
    x_train, y_train, w_init, b_init, tmp_alpha,
    iterations, compute_cost, compute_gradient
)

print(f"\n(w,b) found by gradient descent: ({w_final:8.4f}, {b_final:8.4f})")

```

```python
# Plot cost versus iteration
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 4))

ax1.plot(J_hist[:100])
ax2.plot(1000 + np.arange(len(J_hist[1000:])), J_hist[1000:])

ax1.set_title("Cost vs. iteration (start)")
ax2.set_title("Cost vs. iteration (end)")
ax1.set_ylabel('Cost')
ax2.set_ylabel('Cost')
ax1.set_xlabel('Iteration step')
ax2.set_xlabel('Iteration step')

plt.show()
```

```python
# Make predictions
print(f"1000 sqft house prediction: {w_final * 1.0 + b_final:0.1f} Thousand dollars")
print(f"1200 sqft house prediction: {w_final * 1.2 + b_final:0.1f} Thousand dollars")
print(f"2000 sqft house prediction: {w_final * 2.0 + b_final:0.1f} Thousand dollars")
```

```python
# Visualization function for contour plot with gradient descent path
def plot_contour_with_path(x, y, p_hist, w_range=None, b_range=None):
    """
    Plots cost contours and the gradient descent path.
    """
    # Set default ranges if not provided
    if w_range is None:
        w_range = [0, 400, 5]
    if b_range is None:
        b_range = [-100, 500, 5]
    
    # Create grid for contour plot
    w_vals = np.arange(w_range[0], w_range[1], w_range[2])
    b_vals = np.arange(b_range[0], b_range[1], b_range[2])
    W, B = np.meshgrid(w_vals, b_vals)
    
    # Compute cost for each point in the grid
    Z = np.zeros_like(W)
    for i in range(W.shape[0]):
        for j in range(W.shape[1]):
            Z[i, j] = compute_cost(x, y, W[i, j], B[i, j])
    
    # Create contour plot
    fig, ax = plt.subplots(1, 1, figsize=(12, 6))
    contour = ax.contour(W, B, Z, levels=20, cmap='viridis')
    ax.clabel(contour, inline=True, fontsize=8)
    
    # Plot gradient descent path
    p_hist_arr = np.array(p_hist)
    ax.plot(p_hist_arr[:, 0], p_hist_arr[:, 1], 'ro-', markersize=4, linewidth=1.5, label='GD path')
    ax.plot(p_hist_arr[0, 0], p_hist_arr[0, 1], 'go', markersize=10, label='Start')
    ax.plot(p_hist_arr[-1, 0], p_hist_arr[-1, 1], 'rs', markersize=10, label='End')
    
    ax.set_xlabel('w')
    ax.set_ylabel('b')
    ax.set_title('Cost Contours and Gradient Descent Path')
    ax.legend()
    plt.colorbar(contour, ax=ax, label='Cost')
    plt.show()
```

```python
# Plot with wide view
plot_contour_with_path(x_train, y_train, p_hist)
```

```python
# Plot with zoomed-in view
plot_contour_with_path(
    x_train, y_train, p_hist,
    w_range=[180, 220, 0.5],
    b_range=[80, 120, 0.5]
)
```

```python
# Example: Divergence with Large Learning Rate
# Initialize parameters
w_init = 0
b_init = 0

# Set alpha to a large value to show divergence
iterations = 10
tmp_alpha = 8.0e-1

# Run gradient descent
w_final_div, b_final_div, J_hist_div, p_hist_div = gradient_descent(
    x_train, y_train, w_init, b_init, tmp_alpha,
    iterations, compute_cost, compute_gradient
)
```

```python
# Visualization function for divergence
def plot_divergence(p_hist, J_hist, x, y):
    """
    Plots the divergence behavior with a large learning rate.
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Plot cost history
    ax1.plot(J_hist, 'ro-', markersize=6)
    ax1.set_xlabel('Iteration')
    ax1.set_ylabel('Cost')
    ax1.set_title('Cost vs. Iteration (Diverging)')
    ax1.grid(True)
    
    # Plot parameter path
    p_hist_arr = np.array(p_hist)
    ax2.plot(p_hist_arr[:, 0], p_hist_arr[:, 1], 'bo-', markersize=6)
    ax2.plot(p_hist_arr[0, 0], p_hist_arr[0, 1], 'go', markersize=10, label='Start')
    ax2.set_xlabel('w')
    ax2.set_ylabel('b')
    ax2.set_title('Parameter Path (Diverging)')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()

plot_divergence(p_hist_div, J_hist_div, x_train, y_train)
```
